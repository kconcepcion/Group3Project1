```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Energy = [416, 377, 403]
Electricity = [154, 99, 150]

n=3
r = np.arange(n)
width = 0.25


plt.bar(r, Energy, color = 'b',
		width = width, edgecolor = 'black',
		label='Total Energy')
plt.bar(r + width, Electricity, color = 'g',
		width = width, edgecolor = 'black',
		label='Monthly Electricity')

plt.xlabel("State")
plt.ylabel("Cost ($)")
plt.title("Electricity and Energy Cost 2022")

# plt.grid(linestyle='--')
plt.xticks(r + width/2,['Texas','New Mexico','Arizona'])
plt.legend(loc='upper center')

plt.show()

```
