```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
txoakenergyconsumption=pd.read_csv("EnergyconsumpokTx.csv")
txnmenergyconsumption=pd.read_csv("EnergyconsumpnmTX.csv")
Electricosttxok=pd.read_csv("Txokelctriccost.csv")
Electricosttxnm=pd.read_csv("Txnmelctricost.csv")
```

```python
txoakenergyconsumption
```

```python
#Texas vs Ok energy consumption bar
#Values= [Commercial(trillionBtu), Residential(trillionBtu), GrandTotal]
txoakenergyconsumption1 = txoakenergyconsumption.set_index("Energy Consumption 2020")
barTXOK= txoakenergyconsumption1.plot.bar(color=["b","r"])
plt.xlabel("Types of Consumption",fontsize = 14)
plt.ylabel("BTU",fontsize = 14)
plt.title("OK VS TX Energy Consumption 2020",fontsize = 20)

#plt.xticks(x1,values)
barTXOK
plt.xticks(rotation=45)
           
```

```python
txnmenergyconsumption1 = txnmenergyconsumption.set_index("Energy Consumption 2020")
barTXOK= txoakenergyconsumption1.plot.bar(color=["b","r"])
plt.xlabel("Types of Consumption",fontsize = 14)
plt.ylabel("BTU",fontsize = 14)
plt.title("NM VS TX Energy Consumption 2020",fontsize = 20)

#plt.xticks(x1,values)
barTXOK
plt.xticks(rotation=45)
```
