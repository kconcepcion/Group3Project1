```python
#importing dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
##this scatter plot shows dalls winter freeze temps
winterFreezeTempData = pd.read_csv("project1DFWOKtemps.csv")
winterFreezeTempData
```

```python
#first I am going to read in a csv file, that contains a list of temperatures and dates for dallas and oklahoma city
##to do a comparative analysis

xValue1 = winterFreezeTempData["Date"]
yValue1 = winterFreezeTempData["Texas (DFW)"]
yValue2 = winterFreezeTempData["Oklahoma (Oklahoma City)"]

plt.scatter(xValue1, yValue1, marker="o", facecolors="blue", edgecolors="blue")
plt.xlabel('Dates of Winter Freeze 2021')
plt.ylabel('Temperatures (Fahrenheit) in Dallas, Tx')
plt.title('Dallas Winter Freeze Temperatures')
plt.grid(True)

plt.ylim(-15, 25)
plt.savefig("texasWinterTemp.png")
plt.show()

```

```python
##this scatter plot shows dalls winter freeze temps for oklahoma
yValue2 = winterFreezeTempData["Oklahoma (Oklahoma City)"]

plt.scatter(xValue1, yValue2, marker="o", facecolors="green", edgecolors="green")
plt.xlabel('Dates of Winter Freeze 2021')
plt.ylabel('Temperatures (Fahrenheit) in Oklahoma City, Ok')
plt.title('Oklahoma City Winter Freeze Temperatures')
plt.grid(True)

plt.ylim(-15, 25)
plt.savefig("oklahomaWinterTemp.png")
plt.show()
```

```python
texasTemp1, = plt.plot(xValue1, yValue1, marker="o", color="blue", label="Dallas, TX Temperatures" )
oklahomaTemp1, = plt.plot(xValue1, yValue2, marker="o", color="green", label="Oklahoma City, OK Temperatures")
plt.xlabel('Dates of Winter Freeze 2021')
plt.ylabel('Temperatures (Fahrenheit)')
plt.title('Comparing Texas v Oklahoma Temperatures')
plt.legend(handles=[texasTemp1, oklahomaTemp1], loc="best")
plt.ylim(-15,30)
plt.grid(True)
plt.savefig("texasVoklahomaWinterTemp.png")
```
